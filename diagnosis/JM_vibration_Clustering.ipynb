# 지도학습

```
# 여기서는 여러 알고리즘 사용
from sklearn.linear_model import LogisticRegression # 회귀 모델인데 분류에서 한번 사용
from sklearn.tree import DecisionTreeClassifier # 결정 트리 분류
from sklearn.ensemble import RandomForestClassifier # 랜덤포레스트 분류
from sklearn.naive_bayes import GaussianNB # 가우시안NB (텍스트 처리용)
from sklearn.svm import SVC # 서포트벡터머신 분류
```

```
# 도구
import sklearn
from sklearn.metrics import auc, roc_curve # 곡선하면적으로 알고리즘을 교차로 성능평가
# 데이터를 훈련용과 테스트용으로 나누는 함수
from sklearn.model_selection import train_test_split 
```

```
X_train, X_test, y_train, y_test = train_test_split(Scaled.iloc[:,1:-2], Scaled.iloc[:,-2:-1], test_size=0.2, random_state=1)
```

```
# 알고리즘 준비
als = {
    # 키(알고리즘) : (선의 모양, 알고리즘 객체)
    'LogisticRegression' : ('-', LogisticRegression() ),
    'DecisionTreeClassifier' : ('--', DecisionTreeClassifier(max_depth=5)),
    'RandomForestClassifier' : ('.-', RandomForestClassifier(max_depth=5, max_features=1, n_estimators=10)),
    'GaussianNB' : (':', GaussianNB()),
    #'SVC' : ('-', SVC())
}
```

```
als.items()
```

```
plt.figure(figsize=(10, 10))

for al_name, (lineStyle, model) in als.items():
  # print(al_name, lineStyle, model)
  # 알고리즘 선택
  # 학습
  model.fit(X_train, y_train)
  # 예측
  # 예측 시 확률이 높은 쪽으로 편향되서 예측하는 방법
  pred = model.predict_proba(X_test)
  # 성능평가
  # print(type(pred), pred)
  # 예측값 중 1개만 (특정 긍정비율만) 선택
  pred_s = pd.Series(pred[ : , 1])
  # pred_s
  # roc 커브
  fpr, tpr, _ =roc_curve(y_test, pred_s)
  # 차트 그리기
  plt.plot(fpr, tpr, lineStyle, linewidth=5, label=al_name)
  # 수치로 확인
  print(al_name, auc(fpr, tpr))



plt.legend()
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.plot([0,1], [0,1], 'k--')
plt.show()
```

```
import xgboost as xgb
xgb.__version__
```

```
X_train
```

```
# xgboost를 사용하기 위해서는 데이터를 특정 형식으로 변경해야한다
# DMatrix 형식을 취한다 => 변환된 데이터를 가지고 학습, 예측, 평가
# 하이퍼파라미터 튜닝을 이용하여 최적화
# 사전에 데이터는 75(훈련용 50: 검증용 25):25(테스트용)
dtrain = xgb.DMatrix(data=X_train, label=y_train)
dtest = xgb.DMatrix(data=X_test, label=y_test)
```

```
dab6=xgb.DMatrix(test.iloc[:, 1:-1])
```

```
# 하이퍼파라미터 -> 교차 검증말고 1개씩만 기술, 나중에 추가
'''
 'eval_metric' -> 손실함수, 회귀 : rmse, 분류:error
------------- (회귀시 사용) ---------------
- rmse 
- mse
------------- (분류시 사용) -------------
- logloss
- error : binary classfication error rate
- merror
- miogloss
- auc
'''
# 과적합이 높다 -> eta는 낮추고 max_depth값 낮추고 
# gamma 값 -> L2 정규화값 , alpha값 -> L1 정규화값  값을 올린다
params = {
    'max-_depth' : 3, # 트리의 깊이
    'eta' : 0.1, # 학습시 업데이트값 => learning_rate
    'objective' : 'binary:logistic', # 현재 분류값은 0 or 1
    'eval_metric': 'logloss', # 평가시 사용하는 손실함수, 평가지표
    'early_stopping': 100 # 조기 학습 종료 -> 일반적인 사용은 학습 성과를 평가해서 일정부분 변동이 없으면 종료, 여기서는 최대 쿼터 부여
}
```

```
# 학습시 사용한 데이터 형식 정의
# dtrain 훈련용, dtest 검증용으로 사용
vList = [(dtrain, 'train'), (dtest, 'eval')]
# 학습
# 사용형태가 기존의 머신러닝의 스타일이 아닌 훈련시 학습된 모델을 리턴한다는 개념
xgb_model = xgb.train(params=params, dtrain=dtrain, num_boost_round=400, evals=vList)
```

```
# 상황상 검증 데이터로 사용하지 않은 한번도 접하지 않은 데이터를 테스트
# 현재 데이터가 없어서 그냥 진행
# 데이터는 DMatrix 형태로 변환해서 사용
# 예측값이 0~1 사이로 나와서 (확률값이므로)
# 0.5이상이면 1 아니면 0 이런식으로 분류 처리
pred = xgb_model.predict(dtest)
pred[:10]
```

```
# 예측 확률 처리
preds = [1 if x>0.5 else 0 for x in pred]
preds[:10]
```

```
# 분류 => 혼동행렬, 정확도, 정밀도, 재현율, f1-score, auc, roc_커브
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# 실제값, 예측값
def cal_eval(y_test, pred):
  print('오차행렬(혼동행렬)', confusion_matrix(y_test, pred))
  print('정확도', accuracy_score(y_test, pred))
  print('정밀도', precision_score(y_test, pred))
  print('재현율', recall_score(y_test, pred))
  print('f1_score', f1_score(y_test, pred))
  print('roc, auc', roc_auc_score(y_test, pred))
  pass
```

```
cal_eval(y_test, preds)
```
